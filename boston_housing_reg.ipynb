```python
#Determining outline by calculating z-score.
from scipy import stats
zscore=np.abs(stats.zscore(df))
zscore
```

```python
#Determine the index where zscore is greater than 3
outlierlist=np.where(zscore>3)
print("The indexs of outliers is",outlierlist[0])
print("The number of outliers is",len(outlierlist[0]))
```

```python
#removing the outliers from the dataset
df_1=df[(zscore<3).all(axis=1)]
print("Dimension after removing outliers",df_1.shape)
```

```python
x = df_1.iloc[:,:-1]
y = df_1.iloc[:,-1]
```

```python
#split data
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
if len(x_train) != len(y_train):
    # Handle the inconsistency, either by adjusting the split or addressing missing values
    # For example, you might want to re-split the data
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Now, the number of samples in x_train and y_train should match
# Proceed with model fitting
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#Equation coefficient and intercept of new model
print("Coefficient of independent variables: ")
print(list(zip(model.coef_)))
print("Intercept in the model: ", model.intercept_)
```

```python
#Performance evaluation
df2=pd.DataFrame({"Actual_value":y_test,"Predicted_value":y_pred,"Difference":y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("MAE is: ",mean_absolute_error(y_test,y_pred))
print("MAPE is: ",mean_absolute_percentage_error(y_test,y_pred))
print("MSE is :", mean_squared_error(y_test,y_pred))
print("R2 SCORE is: ",r2_score(y_test,y_pred))
```
